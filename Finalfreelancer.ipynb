```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import json
data=[]
# translate_url = "https://translate.google.com/translate?hl=en&sl=auto&tl=en&u=https://ausschreibungen.giz.de/Satellite/common/project/search.do?method=searchExtended"

# Initialize the WebDriver
driver = webdriver.Chrome()  # Ensure the path to your ChromeDriver is correct if needed
url = "https://ausschreibungen.giz.de/Satellite/common/project/search.do?method=searchExtended"
driver.get(url)  # Load the URL
time.sleep(2)
# Locate the English menu item
Language = driver.find_element(By.XPATH, '//*[@id="mainNav"]/ul[1]/li/a')
actions = ActionChains(driver)
actions.move_to_element(Language).perform()
time.sleep(7)
English_button = driver.find_element(By.XPATH, './/ul[@class="subLevel"]//li//span[@class="languageBlock"]//form//div[@class="controls-float-none"]//input[@id="ENGLISH"]')
English_button.click()
time.sleep(7)

# Function to process text and get the second line if available
def next(text):
    lines = text.split('\n')
    return lines[1] if len(lines) > 1 else ''

def output_json(data, filename='output_data.json'):
    """Outputs the data as a JSON file."""
    # Get the current working directory
    cwd = os.getcwd()
    file_path = os.path.join(cwd, filename)

    # Convert the list of dictionaries to a JSON formatted string
    json_data = json.dumps(data, indent=4)
    
    # Write the JSON string to a file
    with open(file_path, 'w') as file:
        file.write(json_data)
    print(f"Data successfully saved to {filename}")
    

for j in range(8):
    print(f"this is page details---- {j}")
    # Find the table and rows
    table = driver.find_element(By.XPATH, './/table[@class="csx-new-table"]//tbody')
    print("Table found:", table)
    rows = table.find_elements(By.XPATH, './/tr')
    print("Number of rows:", len(rows))
    time.sleep(2)
    for i in range(len(rows)):
        # Extract the publish date
        pub = rows[i].find_element(By.XPATH, './td[1]').text
        sub = rows[i].find_element(By.XPATH, './td[2]').text
        des = rows[i].find_element(By.XPATH, './td[3]').text
        type_ = rows[i].find_element(By.XPATH, './td[4]').text
        adv = rows[i].find_element(By.XPATH, './td[5]').text


        try:
            # Correctly format the XPath with the loop index
            link = rows[i].find_element(By.XPATH, f'//*[@id="masterForm"]/table/tbody/tr[{i + 1}]/td[6]/a/img')
            link.click()
            
            time.sleep(2) # Wait for the new window/tab to load
        
            # Switch to the new window/tab
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(4)
            new_url = driver.current_url
            print("The new URL is:", new_url)
        except Exception as e:
            print("First Link not click, trying second XPath. Error:", e)
            
            # Try the second possible structure of table1 if the first one fails
            try:
                # Correctly format the XPath with the loop index
                link = rows[i].find_element(By.XPATH, f'//*[@id="masterForm"]/table/tbody/tr[{i + 1}]/td[6]/a/img')
                link.click()
                
                time.sleep(2) # Wait for the new window/tab to load
            
                # Switch to the new window/tab
                driver.switch_to.window(driver.window_handles[1])
                time.sleep(4)
                new_url = driver.current_url
                print("The new URL is:", new_url)
    
            except Exception as e:
                print("Second Link also not found. Error:", e)
    
        Token_Id = driver.find_element(By.XPATH, '//*[@id="content"]/div/div[10]/div/div[2]')
        Token_Id=Token_Id.text
        print(Token_Id)
        time.sleep(2)

        if sub=="nv":
            second_page = driver.find_element(By.XPATH, '//*[@id="subNav"]/ul/li[2]/a')
            second_page.click()
            time.sleep(3)
            
            table1 = driver.find_element(By.XPATH, '/html/body/div/div[4]/form/span[2]/div[2]/fieldset/div[2]')
            print("Table found:", table1)
            # Extract information from the second page
            official_name_text = table1.find_element(By.XPATH, '//div[1]//div[@class="controls"]//span[@class="read-only"]').text
            postal_address_text = next(table1.find_element(By.XPATH, '//div[6]').text)
            postal_code_text = next(table1.find_element(By.XPATH, '//div[7]').text)
            town_text = next(table1.find_element(By.XPATH, '//div[8]').text)
            country_text = next(table1.find_element(By.XPATH, '//div[9]').text)
            telephone_text = next(table1.find_element(By.XPATH, '//div[13]').text)
            fax_text = next(table1.find_element(By.XPATH, '//div[14]').text)
            email_text = next(table1.find_element(By.XPATH, '//div[12]').text)
            Internet_address_text = next(table1.find_element(By.XPATH, '//div[4]').text)
        
        
        else:
            second_page = driver.find_element(By.XPATH, '//*[@id="subNav"]/ul/li[3]/a')
            second_page.click()
            time.sleep(3)
            
            # First try to find the first possible structure of table1
            try:
                table1 = driver.find_element(By.XPATH, '/html/body/div[2]/div[4]/div/span[1]/div/div[1]/fieldset/div')
                print("Table found with first XPath:", table1)
        
                # Extract information from the first table structure
                official_name_text = table1.find_element(By.XPATH, '//div[2]//div[@class="controls"]//span[@class="read-only"]').text
                postal_address_text = next(table1.find_element(By.XPATH, '//div[4]').text)
                postal_code_text = next(table1.find_element(By.XPATH, '//div[5]').text)
                town_text = next(table1.find_element(By.XPATH, '//div[6]').text)
                country_text = next(table1.find_element(By.XPATH, '//div[7]').text)
                telephone_text = next(table1.find_element(By.XPATH, '//div[11]').text)
                fax_text = next(table1.find_element(By.XPATH, '//div[12]').text)
                email_text = next(table1.find_element(By.XPATH, '//div[13]').text)
                Internet_address_text = next(table1.find_element(By.XPATH, '//div[16]').text)

            except Exception as e:
                print("First XPath not found, trying second XPath. Error:", e)
                
                # Try the second possible structure of table1 if the first one fails
                try:
                    table1 = driver.find_element(By.XPATH, '/html/body/div/div[4]/form/span[3]/div[2]/fieldset/div[2]')
                    print("Table found with second XPath:", table1)
        
                    # Extract information from the second table structure
                    official_name_text = table1.find_element(By.XPATH, '//div[1]//div[@class="controls"]//span[@class="read-only"]').text
                    postal_address_text = next(table1.find_element(By.XPATH, '//div[6]').text)
                    postal_code_text = next(table1.find_element(By.XPATH, '//div[7]').text)
                    town_text = next(table1.find_element(By.XPATH, '//div[8]').text)
                    country_text = next(table1.find_element(By.XPATH, '//div[9]').text)
                    telephone_text = next(table1.find_element(By.XPATH, '//div[13]').text)
                    fax_text = next(table1.find_element(By.XPATH, '//div[14]').text)
                    email_text = next(table1.find_element(By.XPATH, '//div[12]').text)
                    Internet_address_text = next(table1.find_element(By.XPATH, '//div[4]').text)
        
                except Exception as e:
                    print("Second XPath also not found. Error:", e)

    

        print(official_name_text)
        print(pub)
        print(sub)
        print(des)
        print(type_)
        print(Token_Id)
        print(adv)
        print(postal_address_text)
        print(postal_code_text)
        print(town_text)
        print(country_text)
        print(telephone_text)
        print(fax_text)
        print(email_text)
        print(Internet_address_text)
        time.sleep(5)

        row_data = {
            "Publish_date": pub,
            "Participation_period": sub,
            "Designation": des,
            "Type": type_,
            "Token_id": Token_Id,
            "Advertising_body": adv,
            "official_name": official_name_text,
            "postal_address": postal_address_text ,
            "postal_code": postal_code_text,
            "town": town_text,
            "country": country_text,
            "telephone": telephone_text,
            "fax": fax_text,
            "email": email_text,
            "Internet_address": Internet_address_text,
            "regions": "europe",
        }
        time.sleep(3)
        data.append(row_data) 
        
        
        # Close the new window/tab
        driver.close()
        
        # Switch back to the original window/tab
        driver.switch_to.window(driver.window_handles[0])
          
    print("Click on the next Page")
    time.sleep(2)
    next_page = driver.find_element(By.XPATH, '//*[@id="masterForm"]/table/tfoot/tr/td/div/a[3]')
    next_page.click()
    time.sleep(3)

# Save the scraped data to JSON
output_json(data, 'scraped_data.json')

# Close the driver when done
driver.quit()
```
